## 📊 Loading the Dataset

Here, we load the dataset into a pandas DataFrame to begin analysis.

```python
df = pd.read_csv("Unemployment in India.csv")
df.head()
```

## ⚙️ Data Cleaning

We clean the dataset by removing missing values and formatting columns properly.

```python
df.columns = df.columns.str.strip()
df = df.dropna()
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

## 📌 Dataset Information

We check the structure and statistical summary of the dataset.

```python
print(df.info())
print(df.describe())
```

## 📈 Unemployment Trend Over Time

This visualization shows how unemployment rate changes over time.

```python
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Estimated Unemployment Rate (%)'])
plt.title("Unemployment Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.show()
```

## 🏙️ State-wise Analysis

This graph shows average unemployment rate across different regions.

```python
state_avg = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean()

state_avg.sort_values(ascending=False).plot(kind='bar', figsize=(10,5))
plt.title("Average Unemployment Rate by Region")
plt.xlabel("Region")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.show()
```

## 🦠 Covid-19 Impact Analysis

Here, we compare unemployment rates before and during Covid-19.

```python
df['Year'] = df['Date'].dt.year

before_covid = df[df['Year'] == 2019]['Estimated Unemployment Rate (%)'].mean()
during_covid = df[df['Year'] == 2020]['Estimated Unemployment Rate (%)'].mean()

print("Before Covid (2019):", before_covid)
print("During Covid (2020):", during_covid)
```

## 📅 Monthly Trend Analysis

This shows how unemployment varies across months.

```python
df['Month'] = df['Date'].dt.month

monthly_avg = df.groupby('Month')['Estimated Unemployment Rate (%)'].mean()

monthly_avg.plot(kind='line', marker='o')
plt.title("Monthly Unemployment Trend")
plt.xlabel("Month")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```
